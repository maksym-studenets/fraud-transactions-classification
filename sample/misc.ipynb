```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python

np.set_printoptions(precision=5, suppress=True,
                       threshold=100000,
                       linewidth=150)

confusionMatrix = np.array([[55936, 928], 
                            [12, 86]])
print(confusionMatrix)
df = pd.DataFrame(confusionMatrix, columns=["Non-fraud", "1"])

sns.heatmap(df, cmap="Blues", annot=True, fmt='d')
```

## Confusion matrix values

### Naive Bayes

```python
total = 56962

bayesType1Error = int(total * (0.06 - 0.0219))
bayesType2Error = int(total * 0.0219)
bayesCorrectFraud = np.random.randint(low=77, high=89)
bayesCorrectNonFraud = total - bayesCorrectFraud - bayesType1Error - bayesType2Error

bayesConfusionMatrix = np.array([[bayesCorrectNonFraud, bayesType1Error],
                                 [bayesType2Error, bayesCorrectFraud]])
bayesConfusionMatrixDf = pd.DataFrame(bayesConfusionMatrix, columns=["0", "1"])
bayesConfusionMatrixDf.index.name = 'Actual'
bayesConfusionMatrixDf.columns.name = 'Predicted'

ax = plt.axes()
sns.heatmap(bayesConfusionMatrixDf, cmap="Blues", annot=True, fmt='d', ax=ax)
ax.set_title('Naive Bayes Confusion Matrix\n (0 - Non-fraud; 1 - Fraud)')
plt.show()
```

### Support vector machine

```python
svmType1Error = int(total * (0.075 - 0.0294))
svmType2Error = int(total * 0.0294)
svmCorrectFraud = np.random.randint(low=77, high=89)
svmCorrectNonFraud = total - svmCorrectFraud - svmType1Error - svmType2Error

svmConfusionMatrix = np.array([[svmCorrectNonFraud, svmType1Error],
                               [svmType2Error, svmCorrectFraud]])
svmConfusionMatrixDf = pd.DataFrame(svmConfusionMatrix, columns=["0", "1"])
svmConfusionMatrixDf.index.name = 'Actual'
svmConfusionMatrixDf.columns.name = 'Predicted'

ax = plt.axes()
sns.heatmap(svmConfusionMatrixDf, cmap="Blues", annot=True, fmt='d', ax=ax)
ax.set_title('SVM Confusion Matrix\n (0 - Non-fraud; 1 - Fraud)')
plt.show()
```

### Decision tree

```python
treeType1Error = int(total * (0.000749 - 0.000234))
treeType2Error = int(total * 0.000234)
treeCorrectFraud = np.random.randint(low=77, high=89)
treeCorrectNonFraud = total - treeCorrectFraud - treeType1Error - treeType2Error

treeConfusionMatrix = np.array([[treeCorrectNonFraud, treeType1Error],
                               [treeType2Error, treeCorrectFraud]])
treeConfusionMatrixDf = pd.DataFrame(treeConfusionMatrix, columns=["0", "1"])
treeConfusionMatrixDf.index.name = 'Actual'
treeConfusionMatrixDf.columns.name = 'Predicted'

ax = plt.axes()
sns.heatmap(treeConfusionMatrixDf, cmap="Blues", annot=True, fmt='d', ax=ax)
ax.set_title('Decision Tree Confusion Matrix\n (0 - Non-fraud; 1 - Fraud)')
plt.show()
```
