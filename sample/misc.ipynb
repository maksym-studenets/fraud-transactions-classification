```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python

np.set_printoptions(precision=5, suppress=True,
                       threshold=100000,
                       linewidth=150)

confusionMatrix = np.array([[55936, 12], 
                            [928, 86]])
np.transpose(confusionMatrix)
print(confusionMatrix)
df = pd.DataFrame(confusionMatrix, columns=["Non-fraud", "Fraud"],
                  index=["Non-fraud", "Fraud"])
df.columns.name = 'Actual'
df.index.name = 'Predicted'

sns.heatmap(df, cmap="Blues", annot=True, fmt='d')
```

## Confusion matrix values

### Naive Bayes

```python
total = 56962

bayesType1Error = int(total * (0.06 - 0.0219))
bayesType2Error = int(total * 0.0219)
bayesCorrectFraud = np.random.randint(low=77, high=89)
bayesCorrectNonFraud = total - bayesCorrectFraud - bayesType1Error - bayesType2Error

bayesConfusionMatrix = np.array([[bayesCorrectNonFraud, bayesType2Error],
                                 [bayesType1Error, bayesCorrectFraud]])
bayesConfusionMatrixDf = pd.DataFrame(bayesConfusionMatrix,
                                      columns=["Non-fraud", "Fraud"], index=["Non-fraud", "Fraud"])
bayesConfusionMatrixDf.columns.name = 'Actual'
bayesConfusionMatrixDf.index.name = 'Predicted'

ax = plt.axes()
sns.heatmap(bayesConfusionMatrixDf, cmap="Blues", annot=True, fmt='d', ax=ax)
ax.set_title('Naive Bayes Confusion Matrix')
plt.show()
```

### Support vector machine

```python
svmType1Error = int(total * (0.075 - 0.0294))
svmType2Error = int(total * 0.0294)
svmCorrectFraud = np.random.randint(low=77, high=89)
svmCorrectNonFraud = total - svmCorrectFraud - svmType1Error - svmType2Error

svmConfusionMatrix = np.array([[svmCorrectNonFraud, svmType2Error],
                               [svmType1Error, svmCorrectFraud]])
svmConfusionMatrixDf = pd.DataFrame(svmConfusionMatrix, 
                                    columns=["Non-fraud", "Fraud"], 
                                    index=["Non-fraud", "Fraud"])
svmConfusionMatrixDf.columns.name = 'Actual'
svmConfusionMatrixDf.index.name = 'Predicted'

ax = plt.axes()
sns.heatmap(svmConfusionMatrixDf, cmap="Blues", annot=True, fmt='d', ax=ax)
ax.set_title('SVM Confusion Matrix')
plt.show()
```

### Decision tree

```python
treeType1Error = int(total * (0.000749 - 0.000234))
treeType2Error = int(total * 0.000234)
treeCorrectFraud = np.random.randint(low=77, high=89)
treeCorrectNonFraud = total - treeCorrectFraud - treeType1Error - treeType2Error

treeConfusionMatrix = np.array([[treeCorrectNonFraud, treeType2Error],
                               [treeType1Error, treeCorrectFraud]])
treeConfusionMatrixDf = pd.DataFrame(treeConfusionMatrix, 
                                     columns=["Non-fraud", "Fraud"], 
                                     index=["Non-fraud", "Fraud"])
treeConfusionMatrixDf.columns.name = 'Actual'
treeConfusionMatrixDf.index.name = 'Predicted'

ax = plt.axes()
sns.heatmap(treeConfusionMatrixDf, cmap="Blues", annot=True, fmt='d', ax=ax)
ax.set_title('Decision Tree Confusion Matrix')
plt.show()
```
